```python
# initialize the libraries
import yfinance as yf
import pandas as pd
```

```python
# downlaod the dataset
df = yf.download("AAPL", start = "2020-01-01", end = "2024-01-01")
```

```python
# convert the dataframe to csv and save it locally 
df.to_csv("AAPL.csv")
```

```python
# read the csv and set Date as the index
df = pd.read_csv("AAPL.csv", index_col = "Date")
df
```

```python
# calculating the 20 day returns
df["ret_20"] = df.Close.pct_change(20)
df.head(40)
```

```python
# calculating the signal according to the given condition
df['Signal'] = df['ret_20'] > 0.05
df
```

```python
# drop the NAN rows
df.dropna(inplace = True)
```

```python
# setting the BUY_PRICE column as the next day's Open price
df['Buy_price'] = df.Open.shift(-1)
```

```python
df
```

```python
# the main logic behind Momentum Trading ... checking whether to sell or buy
profits = []
in_position = False

for index, row in df.iterrows():
    if not in_position and row.Signal:
        buy_price = df.loc[index:][0:1].Buy_price.values[0]
        in_position = True
    if in_position:
        sell_price = df.loc[index:][5:6].Close.values[0]
        profit_percent = (sell_price - buy_price) / buy_price
        profits.append(profit_percent)
        in_position = False
```

```python
pd.Series(profits).plot()
```

```python
(pd.Series(profits) + 1).cumprod().plot()
```
